## Нейронные сети: краткая история

В 1943 году нейрофизиолог Уоррен МакКаллох и логик Уолтер Питтс объединились для разработки математической модели искусственного нейрона. В своей [статье](https://link.springer.com/article/10.1007/BF02478259) «Логическое исчисление идей, имманентных нервной деятельности» они заявили, что:

> : Из-за характера нервной деятельности по принципу «все или ничего» нейронные события и отношения между ними можно рассматривать с помощью пропозициональной логики. Установлено, что поведение каждой сети можно описать в этих терминах.

МакКаллох и Питтс поняли, что упрощенную модель реального нейрона можно представить с помощью простого сложения и пороговой обработки, как показано на рис. <<neuron>>. Питтс был самоучкой и к 12 годам получил предложение учиться в Кембриджском университете у великого Бертрана Рассела. Он не принял это приглашение, да и на протяжении всей своей жизни не принимал никаких предложений ученых степеней или руководящих должностей. Большая часть его знаменитых работ была сделана, когда он был бездомным. Несмотря на отсутствие у него официально признанного положения и растущую социальную изоляцию, его работа с Маккаллохом оказала влияние, и ею занялся психолог по имени Фрэнк Розенблатт.

<img alt="Natural and artificial neurons" width="500" caption="Natural and artificial neurons" src="images/chapter7_neuron.png" id="neuron"/>

Розенблатт усовершенствовал искусственный нейрон, чтобы дать ему возможность обучаться. Что еще более важно, он работал над созданием первого устройства, которое действительно использовало эти принципы, — персептрона Mark I. В «Проекте разумного автомата» Розенблатт писал об этой работе: «Сейчас мы станем свидетелями рождения такой машины — машины, способной воспринимать, распознавать и идентифицировать свое окружение без какой-либо подготовки или контроля со стороны человека». Персептрон был построен и смог успешно распознавать простые формы.

Профессор Массачусетского технологического института по имени Марвин Мински (который учился в той же школе на один класс позже Розенблатта!), вместе с Сеймуром Пейпертом написали книгу под названием «Персептроны» (издательство MIT Press) об изобретении Розенблатта. Они показали, что один слой этих устройств не может обучиться некоторым простым, но важным математическим функциям (таким как XOR). В той же книге они также показали, что использование нескольких уровней устройств позволит устранить эти ограничения. К сожалению, только первое из этих открытий получило широкое признание. В результате мировое академическое сообщество почти полностью отказалось от нейронных сетей на следующие два десятилетия.

Возможно, самой ключевой работой в области нейронных сетей за последние 50 лет была многотомная работа «Параллельная распределенная обработка» (PDP) Дэвида Румельхарта, Джеймса Макклеллана и исследовательской группы PDP, выпущенная в 1986 году издательством MIT Press. Глава 1 излагает ту же надежду, что и Розенблатт:

> : Люди умнее современных компьютеров, потому что мозг использует базовую вычислительную архитектуру, которая больше подходит для решения основного аспекта задач естественной обработки информации, с которыми люди так хорошо справляются. ... Мы представим вычислительную основу для моделирования когнитивных процессов, которая кажется… более близкой, чем другие основы, к стилю вычислений, которые может выполнять мозг.

Предпосылка, которую использует здесь PDP, заключается в том, что традиционные компьютерные программы работают совершенно иначе, чем мозг, и, возможно, поэтому компьютерные программы (на тот момент) так плохо справлялись с задачами, которые мозг считал легкими (например, распознавание объектов на картинках). Авторы утверждали, что подход PDP был «ближе
чем другие фреймворки» к тому, как работает мозг, и, следовательно, он может лучше справляться с такого рода задачами.

На самом деле подход, заложенный в PDP, очень похож на подход, используемый в сегодняшних нейронных сетях. В книге параллельная распределенная обработка определяется как требующая:

1. Набор *процессорных блоков*
1. *состояние активации*
1. *Функция вывода* для каждого блока
1. *Схема подключения* между единицами
1. *правило распространения* для распространения паттернов действий через сеть соединений.
1. *Правило активации* для объединения входных данных, воздействующих на единицу, с текущим состоянием этой единицы для создания выходных данных для единицы.
1. *Правило обучения*, согласно которому паттерны связности изменяются с опытом.
1. *Среда*, в которой должна работать система

В этой книге мы увидим, что современные нейронные сети справляются с каждым из этих требований.

В 1980-х годах большинство моделей было построено с использованием второго слоя нейронов, что позволило избежать проблемы, выявленной Мински и Пейпертом (это был их «схема связи между единицами», если использовать приведенную выше схему). И действительно, нейронные сети широко использовались в 80-х и 90-х годах для реальных практических проектов. Однако опять-таки непонимание теоретических вопросов сдерживало поле. Теоретически добавления всего одного дополнительного слоя нейронов было достаточно, чтобы с помощью этих нейронных сетей можно было аппроксимировать любую математическую функцию, но на практике такие сети часто были слишком большими и слишком медленными, чтобы быть полезными.

Хотя 30 лет назад исследователи показали, что для получения практических хороших результатов необходимо использовать еще больше слоев нейронов, только в последнее десятилетие этот принцип получил более широкое признание и применение. Нейронные сети теперь, наконец, реализуют свой потенциал благодаря использованию большего количества слоев в сочетании с возможностью делать это благодаря улучшениям в компьютерном оборудовании, увеличению доступности данных и алгоритмическим настройкам, которые позволяют обучать нейронные сети быстрее и легче. Теперь у нас есть то, что обещал Розенблатт: «машина, способная воспринимать, распознавать и идентифицировать свое окружение без какой-либо подготовки или контроля со стороны человека».

## Как научиться глубокому обучению

Профессор Гарварда Дэвид Перкинс, написавший книгу «Как сделать обучение цельным» (Jossey-Bass), много говорит об обучении. Основная идея состоит в том, чтобы научить *всю игру*. Это означает, что если вы обучаете бейсболу, вы сначала приводите людей на бейсбольный матч или заставляете их играть в него. Вы не учите их тому, как наматывать шпагат, чтобы с нуля сделать бейсбольный мяч, физике параболы или коэффициенту трения мяча о биту.

Пол Локхарт, доктор математики Колумбийского университета, бывший профессор Брауна и учитель математики K-12, воображает в влиятельном [эссе](https://www.maa.org/external_archive/devlin/LockhartsLament.pdf) «Плач математика» кошмарный мир, где музыке и искусству учат так же, как учат математике. Детям не разрешается слушать или играть музыку, пока они не потратят более десяти лет на освоение нотной грамоты и теории, проводя занятия по переносу нот в другую тональность. В художественном классе студенты изучают цвета и аппликаторы, но им не разрешается рисовать до колледжа. Звучит абсурдно? Вот как преподают математику: мы требуем, чтобы студенты годами заучивали их наизусть и изучали сухие, бессвязные *основы*, которые, как мы утверждаем, окупятся позже, спустя много времени после того, как большинство из них бросит этот предмет.

К сожалению, именно здесь начинаются многие учебные ресурсы по глубокому обучению — просят учащихся следовать определению гессиана и теоремам для тейлоровской аппроксимации ваших функций потерь, никогда не приводя примеров реального рабочего кода. Мы не пренебрегаем исчислением. Мы любим исчисление, и Сильвен даже преподавал его в колледже, но мы не думаем, что это лучшее место для начала изучения глубокого обучения!

В глубоком обучении действительно помогает, если у вас есть мотивация исправить свою модель, чтобы заставить ее работать лучше. Именно тогда вы начинаете изучать соответствующую теорию. Но в первую очередь нужно иметь модель. Мы учим практически всему на реальных примерах. По мере создания этих примеров мы погружаемся все глубже и глубже и покажем вам, как делать ваши проекты все лучше и лучше. Это означает, что вы будете постепенно изучать все необходимые теоретические основы в контексте таким образом, чтобы понять, почему это важно и как это работает.

Итак, вот наше обязательство перед вами. На протяжении всей этой книги мы будем следовать следующим принципам:

- Обучение *всей игре*. ### Что такое машинное обучение?

Ваш классификатор представляет собой модель глубокого обучения. Как уже упоминалось, в моделях глубокого обучения используются нейронные сети, которые изначально появились в 1950-х годах и стали мощными совсем недавно благодаря последним достижениям.

Еще одна ключевая часть контекста заключается в том, что глубокое обучение — это просто современная область в более общей дисциплине *машинного обучения*. Чтобы понять суть того, что вы делали, когда обучали собственную модель классификации, вам не нужно разбираться в глубоком обучении. Достаточно увидеть, как ваша модель и ваш процесс обучения являются примерами концепций, применимых к машинному обучению в целом.

Итак, в этом разделе мы опишем, что такое машинное обучение. Мы рассмотрим ключевые концепции и покажем, как их можно проследить до исходного эссе, в котором они были представлены.

*Машинное обучение* — это, как и обычное программирование, способ заставить компьютер выполнять определенную задачу. Но как бы мы использовали обычное программирование, чтобы сделать то, что мы только что сделали в предыдущем разделе: распознать собак и кошек на фотографиях? Мы должны были бы записать для компьютера точные шаги, необходимые для выполнения задачи.

Обычно нам достаточно легко записать шаги для выполнения задачи, когда мы пишем программу. Мы просто думаем о шагах, которые мы бы предприняли, если бы нам пришлось выполнять задачу вручную, а затем переводим их в код. Например, мы можем написать функцию, которая сортирует список. В общем, мы бы написали функцию, которая выглядит примерно так <<basic_program>> (где *входы* могут быть несортированным списком, а *результаты* — отсортированным списком).

````python
#hide_input
#caption A traditional program
#id basic_program
#alt Pipeline inputs, program, results
gv('''program[shape=box3d width=1 height=0.7]
inputs->program->results''')
````

Но для распознавания объектов на фотографии это немного сложно; какие шаги мы предпринимаем, когда распознаем объект на картинке? Мы действительно не знаем, поскольку все это происходит в нашем мозгу без нашего осознания этого!

Еще на заре вычислительной техники, в 1949 году, исследователь IBM по имени Артур Сэмюэл начал работать над другим способом заставить компьютеры выполнять задачи, который он назвал «машинным обучением». В своем классическом эссе 1962 года «Искусственный интеллект: рубеж автоматизации» он писал:

> : Программирование компьютера для таких вычислений является, в лучшем случае, трудной задачей, не в первую очередь из-за присущей самому компьютеру сложности, а, скорее, из-за необходимости прописывать каждую минуту шага процесса в самых раздражающих деталях. . Компьютеры, как скажет вам любой программист, — это гигантские дебилы, а не гигантские мозги.

Его основная идея заключалась в следующем: вместо того, чтобы сообщать компьютеру точные шаги, необходимые для решения проблемы, покажите ему примеры проблемы, которую нужно решить, и пусть он сам придумает, как ее решить. Это оказалось очень эффективным: к 1961 году его программа игры в шашки научилась так многому, что обыграла чемпиона штата Коннектикут! Вот как он описал свою идею (из того же эссе, что и выше):

> : Предположим, мы организуем некоторые автоматические средства проверки эффективности любого текущего назначения веса с точки зрения фактической производительности и обеспечим механизм для изменения назначения веса, чтобы максимизировать производительность. Нам не нужно вдаваться в подробности такой процедуры, чтобы увидеть, что ее можно сделать полностью автоматической, и увидеть, что запрограммированная таким образом машина будет «учиться» на своем опыте.

В этом коротком заявлении заключен ряд мощных концепций:

- Идея «назначения веса»
- Тот факт, что у каждого назначения веса есть некоторая "фактическая производительность"
- Требование наличия «автоматических средств» проверки этой производительности,
- Необходимость в «механизме» (т. е. другом автоматическом процессе) для повышения производительности путем изменения весовых назначений.

Давайте рассмотрим эти концепции одну за другой, чтобы понять, как они сочетаются друг с другом на практике. Во-первых, нам нужно понять, что Самуил имеет в виду под *назначением веса*.

Веса — это просто переменные, а присвоение веса — это конкретный выбор значений для этих переменных. Входные данные программы — это значения, которые она обрабатывает для получения результатов, например, используя пиксели изображения в качестве входных данных и возвращая в результате классификацию «собака». Назначения весов программы — это другие значения, которые определяют, как будет работать программа.

Так как они будут влиять на программу, они в каком-то смысле являются другим видом ввода, поэтому мы обновим нашу базовую картинку в <<basic_program>> и заменим ее <<weight_assignment>>, чтобы принять это во внимание.

````python
#hide_input
#caption A program using weight assignment
#id weight_assignment
gv('''model[shape=box3d width=1 height=0.7]
inputs->model->results; weights->model''')
````

Мы изменили название нашей коробки с *program* на *model*. Это необходимо для того, чтобы следовать современной терминологии и отразить, что *модель* — это программа особого типа: она может делать *много разных вещей*, в зависимости от *весов*. Он может быть реализован множеством различных способов. Например, в программе Сэмюэля для игры в шашки разные значения весов привели бы к разным стратегиям игры в шашки.

(Кстати, то, что Самуэль называл «весами», в наши дни чаще всего называют *параметрами* модели, если вы встречались с этим термином. Термин *веса* зарезервирован для определенного типа параметра модели.)

Затем Сэмюэл сказал, что нам нужны *автоматические средства проверки эффективности любого текущего назначения веса с точки зрения фактической производительности*. В случае с его шашечной программой «фактическая производительность» модели будет заключаться в том, насколько хорошо она играет. И вы можете автоматически протестировать производительность двух моделей, настроив их на игру друг против друга и посмотрев, какая из них обычно побеждает.

Наконец, он говорит, что нам нужен *механизм для изменения назначения весов, чтобы максимизировать производительность*. Например, мы могли бы посмотреть на разницу в весах между выигравшей и проигравшей моделью и немного скорректировать веса в сторону выигрыша.

Теперь мы можем понять, почему он сказал, что такая процедура *может быть сделана полностью автоматической и... машина, запрограммированная таким образом, будет "учиться" на своем опыте*. Обучение стало бы полностью автоматическим, когда корректировка весов также была бы автоматической — когда вместо того, чтобы улучшать модель, корректируя ее веса вручную, мы полагались бы на автоматический механизм, производящий корректировки в зависимости от производительности.

<<training_loop>> показывает полную картину идеи Самуэля об обучении модели машинного обучения.

````python
#hide_input
#caption Training a machine learning model
#id training_loop
#alt The basic training loop
gv('''ordering=in
model[shape=box3d width=1 height=0.7]
inputs->model->results; weights->model; results->performance
performance->weights[constraint=false label=update]''')
````

Обратите внимание на различие между *результатами* модели (например, ходами в игре в шашки) и ее *производительностью* (например, выигрывает ли она игру или как быстро она выигрывает).

Также обратите внимание, что после того, как модель обучена, то есть после того, как мы выбрали наше окончательное, лучшее, любимое назначение весов, мы можем думать о весах как о *части модели*, поскольку мы их не изменяем. более.

Следовательно, на самом деле *использование* модели после ее обучения выглядит как <<using_model>>.

````python
#hide_input
#caption Using a trained model as a program
#id using_model
gv('''model[shape=box3d width=1 height=0.7]
inputs->model->results''')
````

Это выглядит идентично нашей исходной диаграмме в <<basic_program>>, только с заменой слова *program* на *model*. Это важное открытие: *обученную модель можно рассматривать как обычную компьютерную программу*.

> жаргон: машинное обучение: обучение программ, разработанных путем предоставления компьютеру возможности учиться на собственном опыте, а не посредством ручного кодирования отдельных шагов.

### Что такое нейронная сеть?

Нетрудно представить, как могла бы выглядеть модель программы для шашек. Может быть закодирован ряд стратегий шашек и какой-то механизм поиска, а затем веса могут варьироваться в зависимости от того, как выбираются стратегии, на какие части доски обращают внимание во время поиска и так далее. Но совсем не очевидно, как может выглядеть модель для программы распознавания изображений, или для понимания текста, или для многих других интересных задач, которые мы можем себе представить.

Нам бы хотелось, чтобы какая-то функция была настолько гибкой, чтобы ее можно было использовать для решения любой заданной задачи, просто изменяя ее веса. Удивительно, но эта функция действительно существует! Это нейронная сеть, которую мы уже обсуждали. То есть, если рассматривать нейронную сеть как математическую функцию, она оказывается чрезвычайно гибкой функцией в зависимости от своих весов. Математическое доказательство, называемое *универсальной аппроксимационной теоремой*, показывает, что эта функция теоретически может решить любую задачу с любым уровнем точности. Тот факт, что нейронные сети настолько гибки, означает, что на практике они часто являются подходящей моделью, и вы можете сосредоточить свои усилия на процессе их обучения, то есть на поиске хороших весовых коэффициентов.

Но как насчет этого процесса? Можно представить, что вам может понадобиться найти новый «механизм» для автоматического обновления весов для каждой задачи. Это было бы трудоемко. Здесь нам также нужен совершенно общий способ обновления весов нейронной сети, чтобы она улучшалась в любой заданной задаче. Удобно, что это тоже существует!

Это называется *стохастическим градиентным спуском* (SGD). Мы подробно рассмотрим, как работают нейронные сети и SGD, в <<chapter_mnist_basics>>, а также объясним теорему об универсальной аппроксимации. Однако сейчас мы воспользуемся собственными словами Сэмюэля: * Нам не нужно вдаваться в подробности такой процедуры, чтобы увидеть, что ее можно сделать полностью автоматической, и увидеть, что запрограммированная таким образом машина будет «учиться» на своем опыте. *

> Д: Не волнуйтесь, ни SGD, ни нейронные сети не являются математически сложными. Оба почти полностью полагаются на сложение и умножение, чтобы сделать свою работу (но они делают _много_ сложения и умножения!). Основная реакция, которую мы слышим от студентов, когда они видят детали: «И это все?»

Другими словами, напомним, что нейронная сеть — это особый вид модели машинного обучения, которая полностью соответствует первоначальной концепции Сэмюэля. Нейронные сети особенные, потому что они очень гибкие, а это значит, что они могут решать необычайно широкий круг задач, просто находя правильные веса. Это мощно, потому что стохастический градиентный спуск дает нам возможность автоматически находить эти значения веса.

### Ограничения, присущие машинному обучению

На этой картинке мы теперь можем увидеть некоторые фундаментальные вещи об обучении модели глубокого обучения:

- Модель не может быть создана без данных.
- Модель может научиться работать только с шаблонами, наблюдаемыми во входных данных, используемых для ее обучения.
- Этот подход к обучению создает только *прогнозы*, а не рекомендуемые *действия*.
- Недостаточно просто иметь примеры входных данных; нам нужны *метки* и для этих данных (например, изображений собак и кошек недостаточно для обучения модели; нам нужна метка для каждой из них, указывающая, какие из них собаки, а какие кошки).

Вообще говоря, мы видели, что большинство организаций, которые говорят, что у них недостаточно данных, на самом деле имеют в виду, что у них недостаточно *помеченных* данных. Если какая-либо организация заинтересована в том, чтобы сделать что-то на практике с помощью модели, то, по-видимому, у нее есть некоторые исходные данные, на основе которых они планируют запустить свою модель. И, вероятно, какое-то время они делали это каким-то другим способом (например, вручную или с помощью какой-то эвристической программы), поэтому у них есть данные из этих процессов! Например, рентгенологическая практика почти наверняка имеет архив медицинских сканирований (поскольку они должны иметь возможность проверять, как их пациенты прогрессируют с течением времени), но эти сканы могут не иметь структурированных меток, содержащих список диагнозов или вмешательств (поскольку рентгенологи обычно создают отчеты на естественном языке в виде произвольного текста, а не структурированные данные). В этой книге мы будем много обсуждать подходы к маркировке, потому что это очень важный практический вопрос.

Поскольку такие модели машинного обучения могут только делать *предсказания* (т. е. пытаться воспроизвести метки), это может привести к значительному разрыву между целями организации и возможностями модели. Например, в этой книге вы узнаете, как создать *систему рекомендаций*, которая может предсказать, какие продукты может купить пользователь. Это часто используется в электронной коммерции, например, для настройки продуктов, отображаемых на главной странице, путем отображения элементов с наивысшим рейтингом. Но такая модель обычно создается путем изучения пользователя и его истории покупок (*исходные данные*) и того, что они покупали или просматривали (*ярлыки*), а это означает, что модель, скорее всего, расскажет вам о продуктах, которые пользователь уже имеет или уже знает о них, а не о новых продуктах, о которых ему, скорее всего, будет интересно узнать. Это сильно отличается от того, что, скажем, может сделать эксперт в вашем местном магазине книг, когда он задает вопросы, чтобы выяснить ваш вкус, а затем рассказывает вам об авторах или сериях, о которых вы никогда раньше не слышали.

Еще одно важное открытие связано с рассмотрением того, как модель взаимодействует с окружающей средой. Это может создать *петли обратной связи*, как описано здесь:

- Создается *прогнозирующая полицейская модель*, основанная на том, где в прошлом были произведены аресты. На практике это на самом деле не прогнозирование преступлений, а скорее прогнозирование арестов, и поэтому частично просто отражает предвзятость в существующих полицейских процессах.
- Затем сотрудники правоохранительных органов могут использовать эту модель, чтобы решить, на чем сосредоточить свою полицейскую деятельность, что приведет к увеличению количества арестов в этих областях.
- Данные об этих дополнительных арестах затем будут переданы для переобучения будущих версий модели.

Это *цикл положительной обратной связи*, где чем больше используется модель, тем более предвзятыми становятся данные, что делает модель еще более предвзятой и так далее.

Петли обратной связи также могут создавать проблемы в коммерческих условиях. Например, система рекомендаций по видео может быть склонна рекомендовать контент, потребляемый самыми активными зрителями видео (например, теоретики заговора и экстремисты, как правило, смотрят больше видеоконтента в Интернете, чем в среднем), в результате чего эти пользователи увеличивают потребление видео, что приводит к рекомендуется больше таких видео. Более подробно мы рассмотрим эту тему в <<главе_этика>>.

Это два основных типа моделей, которые мы будем исследовать в этой книге. Модель классификации — это модель, которая пытается предсказать класс или категорию. То есть это предсказание на основе ряда дискретных возможностей, таких как «собака» или «кошка». Регрессионная модель — это модель, которая пытается предсказать одну или несколько числовых величин, таких как температура или местоположение. Набор данных Pet содержит 7390 изображений собак и кошек 37 различных пород. Каждое изображение помечено своим именем файла: например, файл *great\_pyrenees\_173.jpg* является 173-м примером изображения собаки породы Великие Пиренеи в наборе данных. Имена файлов начинаются с заглавной буквы, если изображение является кошкой, и строчной буквы в противном случае. Мы должны сообщить fastai, как получить метки из имен файлов, что мы и делаем, вызывая `from_name_func` (что означает, что метки могут быть извлечены с помощью функции, примененной к имени файла) и передавая `is_cat`, которая возвращает `x[0 ].isupper()`, который оценивается как `True`, если первая буква в верхнем регистре (т. е. это кошка).

Самый важный параметр, о котором здесь следует упомянуть, это `valid_pct=0.2`. Это говорит fastai хранить 20% данных и * вообще не использовать их для обучения модели *. Эти 20% данных называются *проверочным набором*; остальные 80% называются *обучающим набором*. Набор проверки используется для измерения точности модели. По умолчанию оставшиеся 20% выбираются случайным образом. Параметр `seed=42` устанавливает *случайное начальное число* в одно и то же значение каждый раз, когда мы запускаем этот код, что означает, что мы получаем один и тот же набор проверки каждый раз, когда мы его запускаем — таким образом, если мы изменим нашу модель и переобучим ее, мы знаем, что любые различия связаны с изменениями в модели, а не с другим набором случайных проверок.

fastai *всегда* покажет вам точность вашей модели, используя *только* проверочный набор, *никогда* обучающий набор. Это очень важно, потому что если вы обучаете достаточно большую модель в течение достаточно долгого времени, она в конечном итоге запомнит метку каждого элемента в вашем наборе данных! Результат на самом деле не будет полезной моделью, потому что нас волнует, насколько хорошо наша модель работает с *ранее невидимыми изображениями*. Это всегда наша цель при создании модели: чтобы она была полезна для данных, которые модель видит только в будущем, после того, как она была обучена.

Даже если ваша модель не полностью запомнила все ваши данные, ранее в процессе обучения она могла запомнить некоторые их части. В результате, чем дольше вы тренируетесь, тем выше будет ваша точность на тренировочном наборе; точность проверочного набора также улучшится на некоторое время, но в конечном итоге она начнет ухудшаться, поскольку модель начнет запоминать обучающий набор, а не находить в данных обобщаемые базовые закономерности. Когда это происходит, мы говорим, что модель *переоснащена*.

Как видите, хотя прогнозы в модели наложения точны для данных, близких к наблюдаемым точкам данных, они далеко за пределами этого диапазона.

**Переоснащение – это самая важная и сложная проблема** при обучении всех специалистов по машинному обучению и всех алгоритмов. Как вы увидите, очень легко создать модель, которая отлично справляется с прогнозированием точных данных, на которых она была обучена, но гораздо сложнее делать точные прогнозы на данных, которые модель никогда раньше не видела. И, конечно же, это те данные, которые действительно будут иметь значение на практике. Например, если вы создадите классификатор рукописных цифр (что мы и сделаем очень скоро!) и будете использовать его для распознавания чисел, написанных на чеках, то вы никогда не увидите ни одного из чисел, на которых обучалась модель — чеки будут слегка различные варианты написания, чтобы иметь дело с. В этой книге вы узнаете много способов избежать переобучения. Однако эти методы следует использовать только после того, как вы подтвердите, что переобучение действительно имеет место (т. е. вы действительно заметили ухудшение точности проверки во время обучения). Мы часто видим, как практикующие специалисты используют методы предотвращения чрезмерной подгонки, даже когда у них достаточно данных, которые им не нужны, и в итоге получают модель, которая может быть менее точной, чем та, которую они могли бы получить.

> важно: набор для проверки: когда вы обучаете модель, вы должны _всегда_ иметь как набор для обучения, так и набор для проверки, и должны измерять точность своей модели только на наборе для проверки. Если вы будете тренироваться слишком долго и с недостаточным количеством данных, вы увидите, что точность вашей модели начнет ухудшаться; это называется _overfitting_. Fastai по умолчанию устанавливает для `valid_pct` значение 0.2, поэтому, даже если вы забудете, fastai создаст для вас проверочный набор!

Пятая строка кода, обучающего наш распознаватель изображений, сообщает fastai создать *сверточную нейронную сеть* (CNN) и указывает, какую *архитектуру* использовать (то есть какую модель создавать), на каких данных мы хотим ее обучать, и какую *метрику* использовать:

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)
```

Почему CNN? Это современный современный подход к созданию моделей компьютерного зрения. В этой книге мы узнаем все о том, как работают CNN. Их структура вдохновлена ​​тем, как работает система человеческого зрения.

В фастае существует множество различных архитектур, которые мы представим в этой книге (а также обсудим, как создать свою собственную). Однако в большинстве случаев выбор архитектуры не является очень важной частью процесса глубокого обучения. Это то, о чем любят говорить ученые, но на практике вряд ли вам нужно тратить на это много времени. Есть несколько стандартных архитектур, которые работают большую часть времени, и в данном случае мы используем _ResNet_, о которой мы будем много говорить в книге; это быстро и точно для многих наборов данных и задач. «34» в «resnet34» относится к количеству уровней в этом варианте архитектуры (другие варианты: «18», «50», «101» и «152»). Модели, использующие архитектуры с большим количеством слоев, требуют больше времени для обучения и более склонны к переоснащению (т. е. вы не можете обучать их столько эпох, прежде чем точность на проверочном наборе начнет ухудшаться). С другой стороны, при использовании большего количества данных они могут быть немного более точными.

Что такое метрика? Метрика — это функция, которая измеряет качество прогнозов модели с использованием проверочного набора и будет выводиться в конце каждой *эпохи*. В этом случае мы используем `error_rate`, функцию, предоставляемую fastai, которая делает именно то, что говорит: сообщает вам, какой процент изображений в проверочном наборе классифицируется неправильно. Другой распространенной метрикой для классификации является «точность» (которая равна «1,0 — частота_ошибок»). fastai предоставляет гораздо больше, которые будут обсуждаться в этой книге.

Понятие метрики может напомнить вам о *потери*, но есть важное отличие. Вся цель потери состоит в том, чтобы определить «меру производительности», которую система обучения может использовать для автоматического обновления весов. Другими словами, хороший выбор для потери — это выбор, который легко использовать для стохастического градиентного спуска. Но метрика предназначена для потребления человеком, поэтому хорошая метрика — это та, которую вам легко понять и которая максимально соответствует тому, что вы хотите, чтобы модель делала. Иногда вы можете решить, что функция потерь является подходящей метрикой, но это не обязательно так.

У `vision_learner` также есть параметр `pretrained`, который по умолчанию имеет значение `True` (поэтому он используется в данном случае, даже если мы его не указали), который устанавливает веса в вашей модели на значения, которые уже были обучены эксперты распознают тысячи различных категорий на 1,3 миллионах фотографий (используя знаменитый набор данных [*ImageNet*](http://www.image-net.org/)). Модель с весами, которые уже были обучены на каком-то другом наборе данных, называется *предварительно обученной моделью*. Вы почти всегда должны использовать предварительно обученную модель, потому что это означает, что ваша модель еще до того, как вы показали ей какие-либо свои данные, уже очень эффективна. И, как вы увидите, в модели глубокого обучения многие из этих возможностей вам понадобятся почти независимо от деталей вашего проекта. Например, части предварительно обученных моделей будут обрабатывать края, градиенты и определение цвета, которые необходимы для многих задач.

При использовании предварительно обученной модели `vision_learner` удалит последний слой, поскольку он всегда специально настраивается для исходной задачи обучения (например, классификации набора данных ImageNet), и заменит его одним или несколькими новыми слоями со случайными весами соответствующего размера. для набора данных, с которым вы работаете. Эта последняя часть модели известна как *голова*.

Использование предварительно обученных моделей — это *наиболее* важный метод, позволяющий нам обучать более точные модели быстрее, с меньшим объемом данных, меньшими затратами времени и денег. Вы можете подумать, что это будет означать, что использование предварительно обученных моделей будет наиболее изученной областью академического глубокого обучения... но вы очень, очень ошибаетесь! Важность предварительно обученных моделей, как правило, не признается и не обсуждается в большинстве курсов, книг или функций библиотеки программного обеспечения и редко рассматривается в научных статьях. Когда мы пишем это в начале 2020 года, все только начинает меняться, но, вероятно, это займет некоторое время. Так что будьте осторожны: большинство людей, с которыми вы разговариваете, вероятно, сильно недооценят то, что вы можете сделать в области глубокого обучения с небольшими ресурсами, потому что они, вероятно, не будут глубоко понимать, как использовать предварительно обученные модели.

Использование предварительно обученной модели для задачи, отличной от той, для которой она была изначально обучена, называется *трансферным обучением*. К сожалению, из-за того, что трансферное обучение недостаточно изучено, лишь немногие области имеют предварительно обученные модели. Например, в настоящее время в медицине доступно несколько предварительно обученных моделей, что затрудняет использование трансферного обучения в этой области. Кроме того, еще недостаточно хорошо понято, как использовать трансферное обучение для таких задач, как анализ временных рядов.

> жаргон: трансферное обучение: использование предварительно обученной модели для задачи, отличной от той, для которой она была изначально обучена.

Шестая строка нашего кода сообщает fastai, как *соответствовать* модели:

```python
learn.fine_tune(1)
```

Как мы уже говорили, архитектура описывает только *шаблон* для математической функции; на самом деле он ничего не делает, пока мы не предоставим значения для миллионов содержащихся в нем параметров.

Это ключ к глубокому обучению — определение того, как подобрать параметры модели, чтобы заставить ее решать вашу проблему. Чтобы соответствовать модели, мы должны предоставить хотя бы одну часть информации: сколько раз смотреть на каждое изображение (известное как количество *эпох*). Количество выбранных вами эпох будет в значительной степени зависеть от того, сколько времени у вас есть в наличии, и от того, сколько времени вам потребуется на практике, чтобы соответствовать вашей модели. Если вы выберете слишком маленькое число, вы всегда сможете обучить большее количество эпох позже.

Но почему метод называется `fine_tune`, а не `fit`? На самом деле у fastai *действительно* есть метод под названием «подгонка», который действительно соответствует модели (т. е. просмотр изображений в тренировочном наборе несколько раз, каждый раз обновляя параметры, чтобы сделать прогнозы все ближе и ближе к целевым меткам). Но в данном случае мы начали с предварительно обученной модели и не хотим выбрасывать все те возможности, которые у нее уже есть. Как вы узнаете из этой книги, есть несколько важных приемов адаптации предварительно обученной модели к новому набору данных — процесс, называемый «тонкой настройкой».

> жаргон: тонкая настройка: метод обучения переносу, при котором параметры предварительно обученной модели обновляются путем обучения в течение дополнительных эпох с использованием задачи, отличной от той, которая использовалась для предварительного обучения.

Когда вы используете метод `fine_tune`, Fastai будет использовать эти приемы за вас. Есть несколько параметров, которые вы можете установить (которые мы обсудим позже), но в форме по умолчанию, показанной здесь, он выполняет два шага:

1. Используйте одну эпоху, чтобы подогнать только те части модели, которые необходимы для правильной работы новой случайной головы с вашим набором данных.
1. Используйте количество эпох, запрошенное при вызове метода, чтобы уместить всю модель, обновляя веса более поздних слоев (особенно головы) быстрее, чем более ранние слои (которые, как мы увидим, обычно не требуют много времени). изменения от предварительно обученных весов).

*Голова* модели — это та часть, которая была добавлена ​​специально для нового набора данных. *Эпоха* — это один полный проход по набору данных. После вызова `fit` результаты после каждой эпохи распечатываются, показывая номер эпохи, потери набора для обучения и проверки («мера производительности», используемая для обучения модели), и любые *метрики*, которые вы запросили (ошибка ставка в данном случае).

Итак, со всем этим кодом наша модель научилась распознавать кошек и собак только из помеченных примеров. Но как оно это сделало?

### Чему научился наш распознаватель изображений

На данном этапе у нас есть распознаватель изображений, который работает очень хорошо, но мы понятия не имеем, что он на самом деле делает! Хотя многие люди жалуются, что глубокое обучение приводит к непроницаемым моделям «черных ящиков» (то есть чему-то, что дает прогнозы, но что никто не может понять), это действительно не может быть дальше от истины. Существует огромное количество исследований, показывающих, как глубоко проверять модели глубокого обучения и получать от них полезные идеи. Сказав это, все виды моделей машинного обучения (включая глубокое обучение и традиционные статистические модели) могут быть сложными для полного понимания, особенно если учесть, как они будут вести себя при столкновении с данными, которые сильно отличаются от данных, используемых для их обучения. Мы будем обсуждать этот вопрос на протяжении всей книги.

### Распознаватель изображений может решать задачи, не связанные с изображениями

Распознаватель изображений может, как следует из его названия, распознавать только изображения. Но многие вещи могут быть представлены в виде изображений, а это значит, что распознаватель изображений может научиться выполнять множество задач.

Например, звук можно преобразовать в спектрограмму, которая представляет собой диаграмму, показывающую количество каждой частоты в каждый момент времени в аудиофайле. Студент Fast.ai Итан Сутин использовал этот подход, чтобы легко превзойти опубликованную точность современного [environmental sound detection model](https://medium.com/@etown/great-results-on-audio-classification-with-fastai-library-ccaf906c5f52) используя набор данных из 8732 городских звуков. `show_batch` Fastai ясно показывает, что каждый отдельный звук имеет довольно характерную спектрограмму, как вы можете видеть на <<img_spect>>.

<img alt="show_batch with spectrograms of sounds" width="400" caption="show_batch with spectrograms of sounds" id="img_spect" src="images/att_00012.png">

Временной ряд можно легко преобразовать в изображение, просто нанеся временной ряд на график. *Машинное обучение* — это дисциплина, в которой мы определяем программу, не полностью написав ее самостоятельно, а изучая данные. *Глубокое обучение* — это специальность машинного обучения, в которой используются *нейронные сети* с несколькими *слоями*. *Классификация изображений* является репрезентативным примером (также известным как *распознавание изображений*). Мы начинаем с *помеченных данных*; то есть набор изображений, где мы присвоили каждому изображению *метку*, указывающую, что оно представляет. Наша цель — создать программу, называемую *моделью*, которая, учитывая новое изображение, сделает точный *предсказание* относительно того, что представляет собой это новое изображение.

Каждая модель начинается с выбора *архитектуры*, общего шаблона того, как такая модель работает внутри. Процесс *обучения* (или *подгонки*) модели — это процесс поиска набора *значений параметров* (или *весов*), которые превращают общую архитектуру в модель, которая хорошо работает для нашего конкретного типа данных. Чтобы определить, насколько хорошо модель работает с одним прогнозом, нам нужно определить * функцию потерь *, которая определяет, как мы оцениваем прогноз как хороший или плохой.

Чтобы ускорить процесс обучения, мы можем начать с *предварительно обученной модели* — модели, которая уже была обучена на чужих данных. Затем мы можем адаптировать его к нашим данным, еще немного обучив его на наших данных, процесс, называемый *точной настройкой*.

Когда мы обучаем модель, ключевой задачей является обеспечение того, чтобы наша модель *обобщала*, то есть извлекала общие уроки из наших данных, которые также применимы к новым элементам, с которыми она столкнется, чтобы она могла делать хорошие прогнозы по этим элементам. . Риск заключается в том, что если мы плохо обучим нашу модель, вместо того, чтобы усвоить общие уроки, она эффективно запомнит то, что уже видела, и тогда она будет делать плохие прогнозы о новых изображениях. Такой отказ называется *переоснащением*. Чтобы избежать этого, мы всегда делим наши данные на две части: *обучающий набор* и *проверочный набор*. Мы обучаем модель, показывая ей только тренировочный набор, а затем оцениваем, насколько хорошо работает модель, видя, насколько хорошо она работает с элементами из проверочного набора. Таким образом, мы проверяем, являются ли уроки, которые модель извлекает из обучающего набора, уроками, которые обобщаются на проверочный набор. Чтобы человек мог оценить, насколько хорошо модель работает на проверочном наборе в целом, мы определяем *метрику*. В процессе обучения, когда модель увидит каждый элемент обучающего набора, мы называем это *эпохой*.

Все эти концепции применимы к машинному обучению в целом. То есть они применяются ко всевозможным схемам определения модели путем ее обучения на данных. Что отличает глубокое обучение, так это особый класс архитектур: архитектуры, основанные на *нейронных сетях*. В частности, такие задачи, как классификация изображений, в значительной степени зависят от сверточных нейронных сетей, о которых мы вскоре поговорим.

## Глубокое обучение подходит не только для классификации изображений

Эффективность глубокого обучения для классификации изображений широко обсуждалась в последние годы, и даже были показаны _сверхчеловеческие_ результаты в сложных задачах, таких как распознавание злокачественных опухолей на компьютерной томографии. Но он может сделать гораздо больше, как мы покажем здесь.

Например, давайте поговорим о том, что критически важно для автономных транспортных средств: локализация объектов на картинке. Если беспилотный автомобиль не знает, где находится пешеход, значит, он не знает, как его объехать! Создание модели, способной распознавать содержимое каждого отдельного пикселя изображения, называется *сегментацией*. Вот как мы можем обучить модель сегментации с помощью fastai, используя подмножество набора данных [*Camvid*](http://www0.cs.ucl.ac.uk/staff/G.Brostow/papers/Brostow_2009-PRL. pdf) из статьи Габрюэля Дж. Бростоу, Жюльена Фокера и Роберто Чиполлы «Семантические классы объектов в видео: база данных достоверных данных высокого разрешения»:

````python
path = untar_data(URLs.CAMVID_TINY)
dls = SegmentationDataLoaders.from_label_func(
    path, bs=8, fnames = get_image_files(path/"images"),
    label_func = lambda o: path/'labels'/f'{o.stem}_P{o.suffix}',
    codes = np.loadtxt(path/'codes.txt', dtype=str)
)

learn = unet_learner(dls, resnet34)
learn.fine_tune(8)
````

Мы даже не будем рассматривать этот код построчно, потому что он почти идентичен нашему предыдущему примеру! (Хотя мы углубимся в модели сегментации в разделе <<chapter_arch_details>>, наряду со всеми другими моделями, которые мы кратко представляем в этой главе, и во многих других.)

Мы можем визуализировать, насколько хорошо она справилась со своей задачей, попросив модель кодировать цветом каждый пиксель изображения. Как видите, он почти идеально классифицирует каждый пиксель в каждом объекте. Например, обратите внимание, что все автомобили окрашены в один и тот же цвет, а все деревья окрашены в один и тот же цвет (в каждой паре изображений левое изображение является меткой истинности, а правое — прогнозом модели). ):

````python
learn.show_results(max_n=6, figsize=(7,8))
````

### Боковая панель: Порядок имеет значение

В блокноте Jupyter очень важен порядок, в котором вы выполняете каждую ячейку. Это не Excel, где все обновляется, как только вы что-то печатаете в любом месте — у него есть внутреннее состояние, которое обновляется каждый раз, когда вы выполняете ячейку. Например, когда вы запускаете первую ячейку записной книжки (с комментарием «CLICK ME»), вы создаете объект под названием «learn», который содержит модель и данные для задачи классификации изображений. Если бы мы запускали ячейку, только что показанную в тексте (ту, которая предсказывает, хороший обзор или нет) сразу после этого, мы получили бы ошибку, поскольку этот объект «обучения» не содержит модели классификации текста. Эта ячейка должна быть запущена после ячейки, содержащей:

```python
from fastai.text.all import *

dls = TextDataLoaders.from_folder(untar_data(URLs.IMDB), valid='test')
learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, 
                                metrics=accuracy)
learn.fine_tune(4, 1e-2)
```

Сами выходные данные могут быть обманчивы, поскольку они включают результаты последнего выполнения ячейки; если вы измените код внутри ячейки, не выполняя его, останутся старые (вводящие в заблуждение) результаты.

За исключением случаев, когда мы прямо упоминаем об этом, блокноты, представленные на [веб-сайте книги](https://book.fast.ai/), предназначены для запуска по порядку, сверху вниз. В общем, при экспериментировании вы обнаружите, что выполняете ячейки в любом порядке, чтобы работать быстро (что является очень удобной функцией Jupyter Notebook), но как только вы изучите и получите окончательную версию своего кода, убедитесь, что вы можете запустить ячейки ваших тетрадей по порядку (ваше будущее не обязательно будет помнить запутанный путь, по которому вы шли иначе!).

В командном режиме двойное нажатие `0` перезапустит *ядро* (которое является двигателем вашего ноутбука). Это очистит ваше состояние и сделает его таким, как будто вы только что начали в блокноте. Выберите «Выполнить все выше» в меню «Ячейка», чтобы запустить все ячейки выше точки, в которой вы находитесь. Мы обнаружили, что это очень полезно при разработке библиотеки fastai.

Давайте перейдем к чему-то гораздо менее привлекательному, но, возможно, значительно более полезному в коммерческом плане: построению моделей из простых *табличных* данных.

> жаргон: табличный: данные в форме таблицы, например, из электронной таблицы, базы данных или CSV-файла. Табличная модель — это модель, которая пытается предсказать один столбец таблицы на основе информации из других столбцов таблицы.

Оказывается, тоже очень похоже. Вот код, необходимый для обучения модели, которая будет предсказывать, является ли человек высокодоходным, на основе его социально-экономического положения:

````python
from fastai.tabular.all import *
path = untar_data(URLs.ADULT_SAMPLE)

dls = TabularDataLoaders.from_csv(path/'adult.csv', path=path, y_names="salary",
    cat_names = ['workclass', 'education', 'marital-status', 'occupation',
                 'relationship', 'race'],
    cont_names = ['age', 'fnlwgt', 'education-num'],
    procs = [Categorify, FillMissing, Normalize])

learn = tabular_learner(dls, metrics=accuracy)
````

Как вы видите, нам пришлось сообщить fastai, какие столбцы являются *категориальными* (то есть содержат значения, которые являются одним из дискретного набора вариантов, например, «оккупация»), а какие *непрерывными* (то есть содержат числовое значение). который представляет собой количество, например «возраст»).

Для этой задачи не существует предварительно обученной модели (как правило, предварительно обученные модели не являются широко доступными для каких-либо задач табличного моделирования, хотя некоторые организации создали их для внутреннего использования), поэтому в данном случае мы не используем `fine_tune`. Вместо этого мы используем `fit_one_cycle`, наиболее часто используемый метод для обучения моделей fastai *с нуля* (т.е. без трансферного обучения):

````python
learn.fit_one_cycle(3)
````

## Наборы проверки и наборы тестов

Как мы уже говорили, цель модели — делать прогнозы относительно данных. Но процесс обучения модели в корне тупой. Если бы мы обучили модель со всеми нашими данными, а затем оценили модель, используя те же данные, мы не смогли бы сказать, насколько хорошо наша модель может работать с данными, которых она не видела. Без этой очень ценной информации, которая поможет нам в обучении нашей модели, есть очень большая вероятность, что она станет хорошо делать прогнозы в отношении этих данных, но будет плохо работать с новыми данными.

Чтобы избежать этого, нашим первым шагом было разделение нашего набора данных на два набора: *тренировочный набор* (который наша модель видит при обучении) и *проверочный набор*, также известный как *набор разработки* (который используется только для оценка). Это позволяет нам проверить, что модель извлекает уроки из обучающих данных, которые обобщаются на новые данные, данные проверки.

Один из способов понять эту ситуацию состоит в том, что в некотором смысле мы не хотим, чтобы наша модель давала хорошие результаты путем «обмана». Если он делает точный прогноз для элемента данных, это должно быть потому, что он изучил характеристики этого типа элемента, а не потому, что модель была сформирована, *фактически увидев этот конкретный элемент*.

Разделение наших проверочных данных означает, что наша модель никогда не увидит их при обучении, поэтому они полностью не запятнаны ими и никоим образом не являются жульничеством. Верно?

На самом деле, не обязательно. Ситуация более тонкая. Это связано с тем, что в реалистичных сценариях мы редко строим модель, просто обучая ее весовые параметры один раз. Вместо этого мы, вероятно, исследуем множество версий модели с помощью различных вариантов моделирования, касающихся сетевой архитектуры, скорости обучения, стратегий увеличения данных и других факторов, которые мы обсудим в следующих главах. Многие из этих вариантов можно описать как выбор *гиперпараметров*. Слово отражает, что они являются параметрами параметров, поскольку они являются выбором более высокого уровня, который определяет значение весовых параметров.

The problem is that even though the ordinary training process is only looking at predictions on the training data when it learns values for the weight parameters, the same is not true of us. We, as modelers, are evaluating the model by looking at predictions on the validation data when we decide to explore new hyperparameter values! So subsequent versions of the model are, indirectly, shaped by us having seen the validation data. Just as the automatic training process is in danger of overfitting the training data, we are in danger of overfitting the validation data through human trial and error and exploration.

The solution to this conundrum is to introduce another level of even more highly reserved data, the *test set*. Just as we hold back the validation data from the training process, we must hold back the test set data even from ourselves. It cannot be used to improve the model; it can only be used to evaluate the model at the very end of our efforts. In effect, we define a hierarchy of cuts of our data, based on how fully we want to hide it from training and modeling processes: training data is fully exposed, the validation data is less exposed, and test data is totally hidden. This hierarchy parallels the different kinds of modeling and evaluation processes themselves—the automatic training process with back propagation, the more manual process of trying different hyper-parameters between training sessions, and the assessment of our final result.

The test and validation sets should have enough data to ensure that you get a good estimate of your accuracy. If you're creating a cat detector, for instance, you generally want at least 30 cats in your validation set. That means that if you have a dataset with thousands of items, using the default 20% validation set size may be more than you need. On the other hand, if you have lots of data, using some of it for validation probably doesn't have any downsides.

Having two levels of "reserved data"—a validation set and a test set, with one level representing data that you are virtually hiding from yourself—may seem a bit extreme. But the reason it is often necessary is because models tend to gravitate toward the simplest way to do good predictions (memorization), and we as fallible humans tend to gravitate toward fooling ourselves about how well our models are performing. The discipline of the test set helps us keep ourselves intellectually honest. That doesn't mean we *always* need a separate test set—if you have very little data, you may need to just have a validation set—but generally it's best to use one if at all possible.

This same discipline can be critical if you intend to hire a third party to perform modeling work on your behalf. A third party might not understand your requirements accurately, or their incentives might even encourage them to misunderstand them. A good test set can greatly mitigate these risks and let you evaluate whether their work solves your actual problem.

To put it bluntly, if you're a senior decision maker in your organization (or you're advising senior decision makers), the most important takeaway is this: if you ensure that you really understand what test and validation sets are and why they're important, then you'll avoid the single biggest source of failures we've seen when organizations decide to use AI. For instance, if you're considering bringing in an external vendor or service, make sure that you hold out some test data that the vendor *never gets to see*. Then *you* check their model on your test data, using a metric that *you* choose based on what actually matters to you in practice, and *you* decide what level of performance is adequate. (It's also a good idea for you to try out some simple baseline yourself, so you know what a really simple model can achieve. Often it'll turn out that your simple model performs just as well as one produced by an external "expert"!)

Помните: ключевым свойством проверочных и тестовых наборов является то, что они должны отражать новые данные, которые вы увидите в будущем. Это может показаться невыполнимым приказом! По определению, вы еще не видели эти данные. Но обычно вы все же знаете некоторые вещи.
